```python
import pandas as pd
from synlab.evaluation import (
    compare_missingness,
    compare_basic_stats,
    compare_category_frequencies,
    compare_marginals,
)

# Load real wine data (pickle format)
real_df = pd.read_pickle("../data/raw/population/dpmm_wine/wine.pkl.gz")

# Load synthetic data generated by the runner
synth_df = pd.read_csv("../outputs/synthetic/population/dpmm_wine_mst_e1/synthetic.csv")

miss = compare_missingness(real_df, synth_df)
stats = compare_basic_stats(real_df, synth_df)
cat_freqs = compare_category_frequencies(real_df, synth_df)
marginals = compare_marginals(real_df, synth_df)

print(miss.head())
print(stats.head())

# Example: inspect one categorical column's frequency table
print(cat_freqs["type"])  # wine type column

# Example: look at a numeric marginal
col_info = marginals["alcohol"]
print(col_info["bins"])
print(col_info["real_prob"])
print(col_info["synthetic_prob"])
```
